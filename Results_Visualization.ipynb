```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load the CSV files and return the dataframe
def load_predictions(file_path):
    df = pd.read_csv(file_path)
    print(f"Loaded data from {file_path}")
    print(f"Available columns: {df.columns}")
    return df

# Function to generate scatter plots and correlation matrices
def visualize_predictions(df, year, model_name):
    actual_column = f'Actual_Total_{year}'
    predicted_column = f'Predicted_Total_{year}'

    # Check if the expected columns are present in the DataFrame
    if actual_column not in df.columns or predicted_column not in df.columns:
        print(f"Error: Missing expected columns {actual_column} or {predicted_column}")
        return

    # Scatter plot for Actual vs. Predicted
    plt.figure(figsize=(8, 6))
    plt.scatter(df[actual_column], df[predicted_column], alpha=0.5, label=f'{model_name} - Predicting {year}')
    plt.plot([df[actual_column].min(), df[actual_column].max()],
             [df[actual_column].min(), df[actual_column].max()], 'r--', lw=2)
    plt.xlabel(f'Actual Total Emissions ({year})')
    plt.ylabel(f'Predicted Total Emissions ({year})')
    plt.title(f'Scatter Plot: Actual vs. Predicted ({model_name} - {year})')
    plt.legend(title='Model')
    plt.show()

    # Correlation matrix for Actual and Predicted
    corr_matrix = df[[actual_column, predicted_column]].corr()
    plt.figure(figsize=(6, 4))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix ({model_name} - {year})')
    plt.show()

# Paths for all model prediction files
file_paths = {
    '2013': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2010 Model/Decision Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2010 Model/Neural_Network_predictions_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2010 Model/Random Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2010 Model/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2016': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2013 mode/Decision_Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2013 mode/Neural Network_predictions2013_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2013 mode/Random_Forest_predictions_with_pollutant_names (1).csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2013 mode/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2019': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2016 Model/Decision Tree_predictions.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2016 Model/Neural_Network_Predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2016 Model/Random Forest_predictions.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2016 Model/XGBoost_predictions.csv',
    },
    '2025': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2019 model/Decision Tree_predictions2019_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2019 model/Neural_Network_test_predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2019 model/Random_Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2019 model/XGBoost_predictions2019_with_pollutant_names.csv',
    }
}

# Iterate through all years and models, load the data, and visualize
for predicted_year, models in file_paths.items():
    for model_name, file_path in models.items():
        df = load_predictions(file_path)
        visualize_predictions(df, predicted_year, model_name)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Function to load the CSV files and return the dataframe
def load_predictions(file_path):
    df = pd.read_csv(file_path)
    print(f"Loaded data from {file_path}")
    return df

# Function to calculate regression metrics
def calculate_metrics(df, year, model_name):
    actual_column = f'Actual_Total_{year}'
    predicted_column = f'Predicted_Total_{year}'

    if actual_column not in df.columns or predicted_column not in df.columns:
        print(f"Error: Missing expected columns {actual_column} or {predicted_column}")
        return None

    # Calculate metrics
    mae = mean_absolute_error(df[actual_column], df[predicted_column])
    rmse = np.sqrt(mean_squared_error(df[actual_column], df[predicted_column]))
    r2 = r2_score(df[actual_column], df[predicted_column])

    return {"Model": model_name, "Year": year, "MAE": mae, "RMSE": rmse, "R": r2}

# Paths for all model prediction files
file_paths = {
    '2013': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2010 Model/Decision Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2010 Model/Neural_Network_predictions_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2010 Model/Random Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2010 Model/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2016': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2013 mode/Decision_Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2013 mode/Neural Network_predictions2013_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2013 mode/Random_Forest_predictions_with_pollutant_names (1).csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2013 mode/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2019': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2016 Model/Decision Tree_predictions.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2016 Model/Neural_Network_Predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2016 Model/Random Forest_predictions.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2016 Model/XGBoost_predictions.csv',
    },
    '2025': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2019 model/Decision Tree_predictions2019_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2019 model/Neural_Network_test_predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2019 model/Random_Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2019 model/XGBoost_predictions2019_with_pollutant_names.csv',
    }
}

# Store model performance results
results = []

# Iterate through all years and models, load the data, calculate metrics
for predicted_year, models in file_paths.items():
    for model_name, file_path in models.items():
        df = load_predictions(file_path)
        metrics = calculate_metrics(df, predicted_year, model_name)
        if metrics:
            results.append(metrics)

# Create a DataFrame to store all results
results_df = pd.DataFrame(results)

# Show the comparison results
print(results_df)

# Plot model comparison based on RMSE and R
plt.figure(figsize=(12, 6))

# RMSE comparison
plt.subplot(1, 2, 1)
sns.barplot(x="Year", y="RMSE", hue="Model", data=results_df)
plt.title('RMSE Comparison Across Models and Years')

# R comparison
plt.subplot(1, 2, 2)
sns.barplot(x="Year", y="R", hue="Model", data=results_df)
plt.title('R Comparison Across Models and Years')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load the CSV files and return the dataframe
def load_predictions(file_path):
    df = pd.read_csv(file_path)
    return df

# Function to gather actual and predicted data for correlation matrix
def gather_data_for_correlation(df, year, model_name):
    actual_column = f'Actual_Total_{year}'
    predicted_column = f'Predicted_Total_{year}'

    if actual_column in df.columns and predicted_column in df.columns:
        # Rename columns to avoid conflicts when merging
        actual_col_name = f'{model_name}_Actual_{year}'
        predicted_col_name = f'{model_name}_Predicted_{year}'
        df = df[[actual_column, predicted_column]].rename(columns={
            actual_column: actual_col_name,
            predicted_column: predicted_col_name
        })
        return df
    else:
        print(f"Error: Missing expected columns {actual_column} or {predicted_column}")
        return None

# Paths for all model prediction files
file_paths = {
    '2013': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2010 Model/Decision Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2010 Model/Neural_Network_predictions_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2010 Model/Random Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2010 Model/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2016': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2013 mode/Decision_Tree_predictions_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2013 mode/Neural Network_predictions2013_with_pollutant_names.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2013 mode/Random_Forest_predictions_with_pollutant_names (1).csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2013 mode/XGBoost_predictions_with_pollutant_names.csv',
    },
    '2019': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2016 Model/Decision Tree_predictions.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2016 Model/Neural_Network_Predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2016 Model/Random Forest_predictions.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2016 Model/XGBoost_predictions.csv',
    },
    '2025': {
        'Decision Tree': '/content/drive/MyDrive/Model Results/2019 model/Decision Tree_predictions2019_with_pollutant_names.csv',
        'Neural Network': '/content/drive/MyDrive/Model Results/2019 model/Neural_Network_test_predictions.csv',
        'Random Forest': '/content/drive/MyDrive/Model Results/2019 model/Random_Forest_predictions_with_pollutant_names.csv',
        'XGBoost': '/content/drive/MyDrive/Model Results/2019 model/XGBoost_predictions2019_with_pollutant_names.csv',
    }
}

# Collect all data for correlation
all_data = pd.DataFrame()

# Iterate through all years and models, load the data, and gather for correlation
for predicted_year, models in file_paths.items():
    for model_name, file_path in models.items():
        df = load_predictions(file_path)
        data = gather_data_for_correlation(df, predicted_year, model_name)
        if data is not None:
            all_data = pd.concat([all_data, data], axis=1)

# Calculate the correlation matrix for all the gathered data
correlation_matrix = all_data.corr()

# Display the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Actual and Predicted Emissions Across All Models and Years')
plt.show()
```
